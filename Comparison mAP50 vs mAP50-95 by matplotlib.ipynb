```python
import matplotlib.pyplot as plt

# Data for YOLOv5 Dataset 1 (bounding box)
categories = ['all', 'pet_dog', 'stray_cat', 'stray_dog']
mAP50_yolov5_dataset1 = [0.939, 0.949, 0.912, 0.955]

# Data for YOLOv5 Dataset 2 (Polygonal)
mAP50_yolov5_dataset2 = [0.958, 0.962, 0.953, 0.909]

# Data for YOLOv8 Dataset 1 (Bounding box)
mAP50_yolov8_dataset1 = [0.934, 0.982, 0.9, 0.92]

# Data for YOLOv8 Dataset 2 (Polygonal)
mAP50_yolov8_dataset2 = [0.931, 0.955, 0.953, 0.909]

# Create a line chart
plt.figure(figsize=(10, 6))

plt.plot(categories, mAP50_yolov5_dataset1, marker='o', label='YOLOv5 Dataset 1 (Bounding box)')
plt.plot(categories, mAP50_yolov5_dataset2, marker='o', label='YOLOv5 Dataset 2 (Polygonal)')
plt.plot(categories, mAP50_yolov8_dataset1, marker='o', label='YOLOv8 Dataset 1 (Bounding box)')
plt.plot(categories, mAP50_yolov8_dataset2, marker='o', label='YOLOv8 Dataset 2 (Polygonal)')

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('mAP50')
plt.title('Comparison of mAP50 for Different Datasets and Architectures')
plt.legend()

# Show the chart
plt.grid(True)
plt.show()
```
